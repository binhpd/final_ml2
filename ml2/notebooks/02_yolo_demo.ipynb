# Notebook 02 - YOLOv11n-seg Demo

Predict + visualize trên dummy data.

```python
import os, sys
os.environ.setdefault('PYTORCH_ENABLE_MPS_FALLBACK', '1')
sys.path.insert(0, '..')
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Pretrained COCO weight (auto-download)
model = YOLO('yolo11n-seg.pt')
print(model.info())
```

```python
# Build dummy data nếu chưa có
from pathlib import Path
if not Path('../data/dummy/images').exists():
    import subprocess
    subprocess.run(['python', '../scripts/build_dummy_data.py', '--n', '20'], check=True)

img_paths = sorted(Path('../data/dummy/images').glob('*.jpg'))[:4]
print(f'Demo trên {len(img_paths)} ảnh')
```

```python
from ml2.yolo_seg.visualize import YOLODocVisualizer
viz = YOLODocVisualizer('yolo11n-seg.pt', device='mps', conf=0.1)

fig, axes = plt.subplots(1, len(img_paths), figsize=(4 * len(img_paths), 4))
for ax, p in zip(axes, img_paths):
    img = cv2.imread(str(p))
    result = viz.visualize(img)
    ax.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
    ax.set_title(p.stem); ax.axis('off')
plt.tight_layout(); plt.show()
```
